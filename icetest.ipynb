```python
newcastle_future = newcastle_future.iloc[:len(newcastle_future)-1]
rotterndam_future = rotterndam_future.iloc[:len(rotterndam_future)-1]
newcastle_historical = newcastle_historical.iloc[:len(newcastle_historical)-1]
rotterndam_historical = rotterndam_historical.iloc[:len(rotterndam_historical)-1]
```

```python
def extract_date(col):
    month_map = {
    "Jan": "01", "Feb": "02", "Mar": "03", "Apr": "04", "May": "05", "Jun": "06",
    "Jul": "07", "Aug": "08", "Sep": "09", "Oct": "10", "Nov": "11", "Dec": "12"
    }
    year_suffix = col.split("(")[-1].split(")")[0][-2:]
    month_str = col.split("(")[-1].split(")")[0][:3]
    
    # Map the month abbreviation to its numerical value
    month = month_map[month_str]
    
    # Add "20" prefix to the year suffix
    year = "20" + year_suffix
    
    # Return formatted date as 'YYYY/MM'
    return f"{year}/{month}"

newcastle_future['date'] = newcastle_future['Contract'].apply(extract_date)
```

```python
newcastle_future.loc[:, 'date'] = newcastle_future['Contract'].apply(extract_date)
rotterndam_future.loc[:, 'date'] = rotterndam_future['Contract'].apply(extract_date)
newcastle_future.loc[:, 'date'] = pd.to_datetime(newcastle_future['date'])
rotterndam_future.loc[:, 'date'] = pd.to_datetime(rotterndam_future['date'])
newcastle_future.loc[:, 'type'] = 'newcastle'
rotterndam_future.loc[:, 'type'] = 'rotterndam'
future_data = pd.concat([newcastle_future, rotterndam_future], axis=0)
```

```python
future_data.to_csv('coal_future_data.csv', index=False)
```
