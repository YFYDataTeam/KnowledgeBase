```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('Weekly_Age_County_Gender_MPXV.csv', encoding='utf8')
```

```python
data
```

```python
gender_data = data.性別.value_counts().reset_index().rename(columns={"性別":"gender"})
```

```python
gender_data
```

```python
import numpy as np

# Creating a DataFrame with the provided data
gender_data = pd.DataFrame({
    'gender': ['M', 'F'],
    'count': [390, 2]
})

# Plotting the bar chart with values on top of the bars
plt.figure(figsize=(5, 4))
bars = plt.bar(gender_data['gender'], gender_data['count'], color='skyblue')
plt.title('Sex Distribution')
plt.xlabel('Sex')
plt.ylabel('Count')
plt.xticks(rotation=0)

# Adding value labels on top of the bars
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval - 10, int(yval), ha='center', va='bottom')

plt.show()
```

```python
age_data = data.年齡層.value_counts().reset_index().rename(columns={"年齡層":"age_level"})
age_data
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Creating the DataFrame with the provided data
age_data = pd.DataFrame({
    'age_level': ['30-34', '35-39', '25-29', '40-44', '20-24', '45-49', '50-54', '55-59', '15-19', '65-69', '4', '70+', '60-64'],
    'count': [123, 93, 65, 52, 32, 12, 5, 3, 3, 1, 1, 1, 1]
})

# Defining the custom order for sorting
age_order = ['4', '15-19', '20-24', '25-29', '30-34', '35-39', '40-44', '45-49', '50-54', '55-59', '60-64', '65-69', '70+']

# Sorting the DataFrame by the custom age order
age_data['age_level'] = pd.Categorical(age_data['age_level'], categories=age_order, ordered=True)
age_data = age_data.sort_values('age_level')

# Plotting the sorted bar chart
plt.figure(figsize=(10, 6))
bars = plt.bar(age_data['age_level'], age_data['count'], color='skyblue')
plt.title('Age Distribution')
plt.xlabel('Age Level')
plt.ylabel('Count')
plt.xticks(rotation=45)

# Adding value labels on top of the bars
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 1, int(yval), ha='center', va='bottom')

plt.show()
```
